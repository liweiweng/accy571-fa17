## Course Goals

Upon completion of this course, students will be expected to understand the basic concepts of data science. Students will learn how to work at a Unix prompt, how to use the Python programming language to process, visualize, and analyze large data sets, and how to perform basic machine learning tasks by using Python.

## Prerequisites

There are no pre-requisites for this course, except for an interest in learning the basic skills necessary for being a data scientist and access to a computer to participate in the course lectures, and to complete the required course assignments, mini-projects, and group project.

Note: At present, we are using NCSA's [Nebula][nb] Openstack cloud computing system to run a course JupyterHub server. Each student is running a Dockerized version of the course software stack. This provides many advantages including robustness against crashes, simplicity of deploying software updates, reduced requirements for students (simply a modern web browser, we have used tablets and smart phones), and simplifying assignment and mini-project submission.

## Academic Integrity

Academic honesty is essential to this course and the University. Any instance of academic dishonesty (including but not limited to cheating, plagiarism, falsification of data, and alteration of grades) will be documented in the student's academic file. In addition, **at a minimum** the particular assessment, quiz, assignment, mini-project, or group project will be given a zero. Serious or repeated offenses may be punished more severely.

**Guidelines for collaborative work**: Discussing course material with your classmates is in general a good idea, but each student is expected to do his or her own work. On assignments or mini-projects, you may discuss the problems and concepts behind them, but you are responsible for your own answers. Please do not post code in the forums! Finally, on assessments and quizzes, your answers must of course be your own. For further info, see the [Student Code, Part 4. Academic Integrity](http://studentcode.illinois.edu/article1_part4_1-401.html).

**Automatic Plagiarism Detection**
This semester we will employ an automatic plagiarism detection tool on each assignment. This tool easily identifies submissions that are more similar than would be randomly expected. Each week the instructional staff will review the output from this tool, and as necessary formally identify students who we believe are guilty of cheating. At a minimum, the following policy will be enforced:

- **First** Offense: Warning.
- **Second** Offense: Zero on affected assignment.
- **Third** Offense: Final grade reduced by one letter grade and a notification sent to home unit.
- **Fourth** Offense: Automatic **F** in course and a notification sent to the Dean of Students.

Furthermore, each argument about the enforcement of this policy that is deemed frivolous by the instructional staff will result in ten percent reduction in the affected assignment score.

## Weekly Format

Each week will contain two modules, which will each provide learning objectives and an outline of the activities for that week with a list of all deadlines and corresponding point values for assignments.

### Videos

Each week there will be one instructor created overview video that will offer a broader context for the new week, explain key concepts, and demonstrate important tasks. In addition, there will be per-lesson videos that go into more depth on the readings and notebooks associated with each lesson. These videos will be linked from the relevant course module or lesson overview notebook on the JupyterHub server. 

To view the instructor videos, you will need to login to the Illinois Mediaspace. You will not be directly graded for watching the videos; however, we strongly encourage you to watch them as they contain information and demonstrations that will prove helpful as you work through each week's lessons. Note that information presented in course videos may be relevant for quizzes and assignments.

### Readings

Readings will consist of articles and excerpts from books and Web sites, internet-accessible videos demonstrating a concept, and, in some cases, IPython notebooks that can be viewed by interacting with them via the course JupyterHub server. You will be required to read and be familiar with the content of these documents. Readings are contextualized as part of the weekly lesson content and are located in the "Readings" section of each lesson. For those wishing to obtain a deeper understanding, we also present optional readings that provide additional information that is not required for either course quizzes or assignments.

### Lessons 

Lessons will expand upon, or clarify key concepts in the reading assignments or supplement or add to the reading. All lessons for a given week must be completed by 6:00 PM Central on Thursday of that week.

### Assignments

Every module contains an assignment that will involve one or more computational tasks related to the focus for that given module. Your submitted assignments will be collected automatically at the deadline from the course JupyterHub server. These assignments will be automatically graded.

### Mini-Projects

Every module but the first and last contains a mini-project that will involve one or more computational tasks related to the focus for that given module. Your submitted mini-projects will be collected at the deadline from the course JupyterHub server. These mini-projects will be graded by the instructional staff.

You also will be given up to five mini-projects from other students in the course to grade as a module mini-project peer assessment. You will receive fifteen points for simply grading your peer's assignments. You do not receive a grade from your peers, this is simply meant to be informative to both the reviewer (who sees how other students solved the mini-project) and to the submitter who gets feedback from fellow students on their submission.

To receive full credit from instructor grading, your mini-project must be submitted prior to the deadline. There will be ***NO*** grace period, late mini-projects will not be accepted. The mini-project deadline is 6:00 PM Central on the Monday or Tuesday (depending on the module) following the relevant week.

** Note: In Fall 2017, Mini-Projects were only assigned for Modules 1-4.**

### Group Projects

You will work in a group on one large data analytics task. The groups will be assigned by the instructional staff and the assigned group project will contain full details of the task. Your submitted notebook will be graded out of 600 total points, the in-class presentation will be worth a maximum of 300 points, and you will review your fellow group-mates for a maximum of 100 points. Thus, the group project is worth a total of 1000 points.

### Peer Review 

Module mini-projects will be reviewed by your course peers, as well as instructor grading. You will not be given points from your peers! 135 points (out of the maximum 150 points for each mini-project) are assigned from  instructor review. You will receive 15 points each week for simply viewing and grading your peers' mini-projects. Note that you can (and should) still grade your peers even if you miss an mini-project submission. Peer review of an mini-project must be completed by 6:00 PM Central on Friday of the following week (i.e., you submit your mini-project on a Monday and you must peer assess other students mini-projects by the following Friday). You will be assigned several mini-projects to grade at approximately 12:00 pm, Wednesday of each week (and perhaps earlier).

| **Item**              | **Grade**      |
| --------------------- | -------------- |
| Instructor Assessment | 135 points     |
| Peer Grading          | 15 points      |
| Peer Assessments      | 0 points      |
| **Total**             | **150** points |

Several final comments about peer grading:

1. We employ peer grading in large part to enable you to see other approaches to solving the mini-projects. Treat this as an important aspect of the course and give reasonable grades and provide feedback to your fellow students.

2. The instructional staff will review peer comments to ensure that peer grading is being done appropriately, which will include verification of beneficial comments.

3. Note that you do not actually receive a grade from your peers. Just feedback on how others view your mini-projects solution. Treat this as valuable insight into how to improve your work.

### Weekly Quizzes

Each week will conclude with a weekly quiz. The weekly quiz is designed to test your overall mastery of the content for each given week. Unlike the lesson quizzes, weekly quizzes will be timed and will not allow multiple attempts. The quiz must be completed by 6:00 PM Central on Friday of that week.

### Exams 

This course is project-based in its use of assignments, mini-projects, and a group project that build progressively on content mastery, application, and peer review; there are **no exams** in this course.
